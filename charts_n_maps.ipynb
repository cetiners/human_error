```python
def print_coords(map, meta):

    coords = map.meta[meta]

    x = [i[0] for i in coords]
    y = [i[1] for i in coords]

    fig, ax = plt.subplots()
    fig.set_dpi(150)
    fig.set_size_inches(8,8)

    ax.scatter(x,y,c = "black",s=1)
    ax.set_title(meta.title().replace("_"," "), fontsize=12)

def print_maps(map, meta):

    coords = map.meta[meta]
    with plt.style.context("seaborn- whitegrid"):
        plt.rcParams["axes.grid"] = True
        fig, ax = plt.subplots()
        fig.set_dpi(150)
        fig.set_size_inches(8,8)
        ax.imshow(coords.T, alpha=0)
        ax.set_title(meta.title().replace("_"," "), fontsize=12)




```

```python
with open("logs/map_nw", "rb") as f:
    nw = pickle.load(f)

nw.land_mask()
```

```python
## INITAL MAP POINTS

print_coords(nw, "initial_centroids")

print_coords(nw, "relaxed_centroids")

print_maps(nw, "unblurred")

print_maps(nw, "final")

print_maps(nw, "sea_mask")
```
